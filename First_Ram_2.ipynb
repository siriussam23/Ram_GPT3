```python
txtPDF1 = ""
txtPDF2 = ""

PDFpath1 = "/Users/scrysam/Movies/Santosh_P/iSpatialTec/Ram/Data_Standards_Compliance_and_Maturity_Dashboard_application_enhancements/Data_Management_Policy.pdf"
PDFpath2 = "/Users/scrysam/Movies/Santosh_P/iSpatialTec/Ram/Data_Standards_Compliance_and_Maturity_Dashboard_application_enhancements/Open_Data_Policy_V3.pdf"

pdfFile1 = open(PDFpath1 , 'rb')
pdfFile2 = open(PDFpath2 , 'rb')

pdfReader1= PyPDF2.PdfFileReader(pdfFile1)
pdfReader2 = PyPDF2.PdfFileReader(pdfFile2)

pageno1 = (pdfReader1.numPages) 
pageno2 = (pdfReader2.numPages) 

for x in range(0,pageno1):
    page1 = pdfReader1.getPage(x)
    sa = page1.extractText()
    txtPDF1 = txtPDF1 + "\n" +sa
print (txtPDF1)

for x in range(0,pageno2):
    page2 = pdfReader2.getPage(x)
    sa = page2.extractText()
    #print (sa[0:100])
    #print (sa)
    txtPDF2 = txtPDF2 + "\n" +sa
print (txtPDF2)

pdfFile1.close()
pdfFile2.close() 
```
